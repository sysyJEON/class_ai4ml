소방신고별 신고시각과 출동시각 (기출 6회)

신고일자 (yyyymmdd) / 신고시각 (HHMMSS , 00시 , 00분의 경우 등 생략)

````python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e6_p1_1.csv')
df.head(5)
````

````python
# 각 구급 보고서 별 출동시각과 신고시각의 차이를 ‘소요시간’ 컬럼을 만들고
# 초(sec)단위로 구하고 소방서명 별 소요시간의 평균을 오름차순으로 정렬 했을때
# 3번째로 작은 소요시간의 값과 소방서명을 출력하라

df['소요시간'] = (
    pd.to_datetime(df['출동일자'].astype('str') + df['출동시각'].astype('str').str.zfill(6))
    - pd.to_datetime(df['신고일자'].astype('str') + df['신고시각'].astype('str').str.zfill(6))
).dt.total_seconds()

ans = df.groupby(['소방서명'])['소요시간'].mean().sort_values().reset_index().iloc[2].values
print(ans)
````

## loc 문제

보스턴집값, 각 행은 지역구별 집값관련된 메타정보 (기출 2회)

````python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e2_p1_1.csv')
df.head(2)
````

````python
# CRIM값이 가장 큰 10개의 지역을 구하고
# 10개의 지역의 CRIM값을 그 중 가장 작은 값으로 대체하라.
# AGE 컬럼 값이 80이상인 대체 된 CRIM 평균값을 구하라.
df = df.sort_values('CRIM', ascending=False).reset_index(drop=True)
df.loc[:9, 'CRIM'] = df.loc[:9, 'CRIM'].min()
ans = df[df['AGE']>=80]['CRIM'].mean()
print(ans)
````

````python
# DIS 평균으로 부터 1.5 * 표준편차를 벗어나는 영역을 이상치라고 판단하고
# DIS 컬럼의 이상치들의 합을 구하여라.

mean = df['DIS'].mean()
sigma = df['DIS'].std()
ans = df.loc[(df['DIS']>mean+1.5*sigma) | (df['DIS']<mean-1.5*sigma), 'DIS'].sum()
print(ans)
````

종량제 봉투 가격 데이터 (기출 5회)

````python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p1_1_.csv')
df.head(5)
````

````python
# 20L가격과 5L가격이 모두 0원이 아닌 데이터만 필터를 한 후,
# 각 row별로 20L가격과 5L가격의 차이를 ‘차이가격’ 이라 부른다고 하자.
# 시도명 별 차이가격의 평균가격을 비교할때
# 그 값이 가장 큰 금액을 반올림하여 소숫점 이하 1자리까지 구하여라

new_df = df.loc[(df['20L가격']!=0) & (df['5L가격']!=0)]
new_df['차이가격'] = new_df['20L가격'] - new_df['5L가격']
diff = new_df.groupby(['시도명'])['차이가격'].mean()
max = diff.max()
ans = round(max, 1)
print(ans)
````

## 분류 문제(기출 4회)

- Segmentation의 클래스 예측
- 평가 지표 f1 score

````python
# 1. 사용 모듈 불러오기
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score

# 2. 데이터 불러오기
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e4_p2_train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e4_p2_test.csv')

display(train.head(2))
display(test.head(2))

# 3. 전처리
# 1) 결측치 처리
# 결측치 확인
display(train.isnull().sum())
display(test.isnull().sum())

# 결측치 채우기
train['Ever_Married'] = train['Ever_Married'].fillna(train['Ever_Married'].mode()[0])
train['Graduated'] = train['Graduated'].fillna(train['Graduated'].mode()[0])
train['Profession'] = train['Profession'].fillna(train['Profession'].mode()[0])
train['Work_Experience'] = train['Work_Experience'].fillna(train['Work_Experience'].mode()[0])
train['Family_Size'] = train['Family_Size'].fillna(train['Family_Size'].mode()[0])
train['Var_1'] = train['Var_1'].fillna(train['Var_1'].mode()[0])

test['Ever_Married'] = test['Ever_Married'].fillna(test['Ever_Married'].mode()[0])
test['Graduated'] = test['Graduated'].fillna(test['Graduated'].mode()[0])
test['Profession'] = test['Profession'].fillna(test['Profession'].mode()[0])
test['Work_Experience'] = test['Work_Experience'].fillna(test['Work_Experience'].mode()[0])
test['Family_Size'] = test['Family_Size'].fillna(test['Family_Size'].mode()[0])
test['Var_1'] = test['Var_1'].fillna(test['Var_1'].mode()[0])

# 결측치 재확인
display(train.isnull().sum())
display(test.isnull().sum())

# 2) 범주형 변경
# Family_Size 범주형 변경
train['Family_Size'] = train['Family_Size'].astype(str)
test['Family_Size'] = test['Family_Size'].astype(str)

# 3) 변수 생성
x = train.drop(columns = ['ID', 'Segmentation'])
y = train['Segmentation']

test = test.drop(columns = ['ID'])

x_dum = pd.get_dummies(x)
test_dum = pd.get_dummies(test)
test_dum = test_dum.reindex(columns = x_dum.columns, fill_value = 0)
# train에는 있고, test에는 없는 데이터 존재 -> 컬럼 숫자가 다름 -> reindex

# 4. 모델 적용
x_train, x_val, y_train, y_val = train_test_split(x_dum, y, test_size=0.33, random_state=42)

rf = RandomForestClassifier(random_state=23)
rf.fit(x_train, y_train)

# 5. 예측 및 결과 저장
pred_val = rf.predict(x_val)
print('validation test f1 score: ', f1_score(y_val, pred_val, average='macro'))

pred_test = rf.predict(test_dum)
submission = pd.DataFrame({'pred' : pred_test})
submission.to_csv('result1.csv', index=False)
````

## 회귀 문제(기출 5회)

````python
# 1. 사용 모듈 불러오기
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# 2. 데이터 불러오기
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p2_train_.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p2_test_.csv')

display(train.head(2))
display(test.head(2))

# 3. 전처리
# 1) 결측치 처리
# display(train.isnull().sum())
# display(test.isnull().sum())
# 결측치 없음

# 2) 변수 생성
x = train.drop(columns = ['ID', 'price'])
y = train['price']

test = test.drop(columns = ['ID'])

x_dum = pd.get_dummies(x)
test_dum = pd.get_dummies(test)
test_dum = test_dum.reindex(columns = x_dum.columns, fill_value=0)
# train에는 있고, test에는 없는 데이터 존재 -> 컬럼 숫자가 다름 -> reindex

# 4. 모델 적용
x_train, x_val, y_train, y_val = train_test_split(x_dum, y, test_size=0.33, random_state=42)

rf = RandomForestRegressor(random_state=23)
rf.fit(x_train, y_train)

# 5. 예측 및 결과 저장
pred_val = rf.predict(x_val)
print('validation test rmse: ', np.sqrt(mean_squared_error(y_val, pred_val)))

pred_test = rf.predict(test_dum)
submission = pd.DataFrame({'pred' : pred_test})
submission.to_csv('result2.csv', index=False)
````

## 카이제곱 검정(기출 3회)

다이어트약의 전후 체중 변화 기록이다.

투약 후 체중에서 투약 전 체중을 뺏을 때 값은 일반 적으로 세가지 등급으로 나눈다.

-3이하 : A등급, -3초과 0이하 : B등급, 0 초과 : C등급.

약 실험에서 A,B,C 그룹간의 인원 수 비율은 2:1:1로 알려져 있다.

위 데이터 표본은 각 범주의 비율에 적합한지 카이제곱 검정하려한다.

````python
import pandas as pd
df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e3_p3_1.csv')
df.head()
````

````python
# 1. A등급에 해당하는 유저는 몇명인지 확인하라
df['dels'] = df['투약후'] - df['투약전']
def filter(x):
  if x <= -3:
    return 'A'
  elif x <= 0:
    return 'B'
  else:
    return 'C'
df['filter'] = df['dels'].apply(filter)
ans1 = df['filter'].value_counts().sort_index()['A']
print(ans1)

# 2. 카이제곱검정 통계량을 반올림하여 소숫점 이하 3째자리까지 구하여라
from scipy.stats import chisquare
observed = df['filter'].value_counts().sort_index().values
expected = [len(df)*(2/4), len(df)*1/4, len(df)*(1/4)]
s, p = chisquare(observed, f_exp=expected)
print(round(s, 3))

# 3. 카이제곱 검정 p값을 반올림하여 소숫점 이하 3자리까지 구하고,
# 유의수준 0.05하에서 귀무가설과 대립가설중 유의한 가설을 하나를 선택하시오(귀무/대립)
round_p = round(p,3)
print(round_p) # >0.05 귀무, <0.05 대립
````

## 카이제곱 독립성 검정(기출 4회)

어느 대학교의 신입생의 학과별 성별에 대한 데이터이다.

학생들의 학과와 성별이 서로 독립적인지 여부를 확인하기 위해 카이제곱 독립성 검정을 실시 하려한다.

````python
import pandas as pd
df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e4_p3_1_.csv')
df.head()
````

````python
# 1. 학과 평균 인원에 대한 값을 소숫점 이하 3자리까지 구하여라
ans1 = round(df.groupby(['학과']).size().mean(), 3)
print(ans1)

# 2. 카이제곱검정 독립성 검정 통계량을 소숫점 이하 3자리까지 구하여라
from scipy.stats import chi2_contingency
crosstab = pd.crosstab(df['학과'], df['성별'])
s, p, dof, expected = chi2_contingency(crosstab)
ans2 = round(s, 3)
print(ans2)

# 3. 카이제곱검정 독립성 검정의 pvalue를 소숫점 이하 3자리까지 구하여라
# 유의수준 0.05하에서 귀무가설과 대립가설중 유의한 것을 출력하라
ans3 = round(p, 3)
print(ans3)
````

## 카이제곱 독립성 검정(기출 6회)

A 도시의 남성 600명과 여성 550명이 있다.

남성들 중 흡연자 비율은 0.2이며 여성들 중 흡연자 비율은 0.26이다.

남성과 여성 간에 흡연 여부에 따른 인구 비율이 다른지 확인하고 싶다.

````python
# 1. 유의 수준 0.05하 귀무가설에 대해 기각 / 채택 여부와 p-value값을 각각 출력하라
import numpy as np
from scipy.stats import chi2_contingency
smoking_male = int(600*0.2)
smoking_female = int(550*0.26)
non_smoking_male = 600 - smoking_male
non_smoking_female = 550 - smoking_female
data = np.array([[smoking_male, non_smoking_male], [smoking_female, non_smoking_female]])
s, p, dof, expected = chi2_contingency(data)
print(p)
````
